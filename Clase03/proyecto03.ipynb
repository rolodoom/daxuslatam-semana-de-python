# Análisis de datos

```python
data["tienda"].value_counts().to_frame()
```

```python
data["ciudad"].value_counts().to_frame()
```

```python
data["forma_pago"].value_counts().to_frame()
```

```python
data.groupby("tienda")["precio"].sum().to_frame()
```

```python
data.groupby("forma_pago")["precio"].sum().to_frame()
```

```python
datos_agrupados = data.groupby(["tienda","forma_pago"])["precio"].sum().to_frame()
```

```python
datos_agrupados.to_excel("facturacion.xlsx")
```

## Visualización de Datos

```python
import plotly.express as px
```

```python
px.histogram(data, x="tienda")
```

```python
grafico = px.histogram(data, x = "tienda", text_auto=True, title="Facturación", color="forma_pago" )
grafico.show()
grafico.write_html("facturacion.html")
```
